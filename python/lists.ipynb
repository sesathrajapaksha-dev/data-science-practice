ordered, mutable (changeable) collection---python arrays can hold multiple datatypes

```python
scores = [10,20,30,40,50]
mixed = [1,"two",3.0,True]
print(scores[0])
print(scores[-1])
```

```python
print(scores[0:3])  # 0 1 2 
print(scores[:3])  # beginning to 3 (not including 3)
print(scores[2:]) # 2 to all the way end
print(scores[::2]) # all , but ekak ara ekak
print(scores[::-1]) # reversed
```

```python
scores.append(100) # add 100 to end
scores.insert(1,88) # insert 88 at index 1, push others aside
print(scores)
scores.remove(20)  # remove first occurence of value 78
removed = scores.pop()  # removes and return last item
scores[0] = 99   # overwrite index0
print(len(scores)) 
```

```python
for i,score in enumerate(scores):
    print(i,score)
```
