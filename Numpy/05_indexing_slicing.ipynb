```python
import numpy as np
```

```python
a=np.array([10,20,30,40,50,60,70,80])
print(a[0])
print(a[3])
print(a[-1])
```

```python
print(a[0:5:2])
print(a[::2])
print(a[::-1])
```

```python
b=[[1,2,3],
   [4,5,6],
   [7,8,9]]
b=np.array(b)
print(b[0,0])
print(b[1,2])
print(b[2,-1])
```

```python
print(b[0,])
print(b[:,2])
print(b[0:2,0:2])
print("----")
print(b[:,1:])
```
