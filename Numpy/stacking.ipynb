```python
import numpy as np
a = np.array([10, 20, 30, 40, 50, 60])
print(a[[0,2,3,5]])
print(a[[3,3,0]])

b = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
print(b[[0,2]])

print(b[[0,1,2],[0,1,0]])
print(b[[0,2],[1,0]])


```

```python
# Stacking
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])

print(np.vstack((a,b)))
print("-----------")
print(np.hstack((a,b)))

c = np.array([[1, 2], 
              [3, 4]])

d = np.array([[5, 6], 
              [7, 8]])

print(np.concatenate((c,d),axis=0))
print(np.concatenate((c,d),axis=1))

e=np.array([1,2,3,4,5,6])
f=np.split(e,3)
print(f[0])
```
