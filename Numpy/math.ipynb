```python
import numpy as np

a = np.array([1, 4, 9, 16])
b = np.array([-3, -1, 2, 5])

print(np.sqrt(a))
print(np.exp(b))
print(np.log(a))
print(np.abs(b))
```

```python
scores=np.array([70,95,60,88,75])
print(np.sort(scores))

print(np.argsort(scores))
print(np.where(scores>80))
print(np.argmax(scores))
print(np.argmin(scores))
```
