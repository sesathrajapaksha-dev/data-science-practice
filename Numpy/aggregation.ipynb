```python
## Masking and Filtering
```

```python
scores = np.array([55, 90, 40, 78, 65, 88, 30])
```

```python
mask = scores >60
print(mask)
```

```python
passing_scores =scores[mask]
print(passing_scores)
mask2=passing_scores >85
print(passing_scores[mask2])
```

```python
print(scores[scores>60])
```
