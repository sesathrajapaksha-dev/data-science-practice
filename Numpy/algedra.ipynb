```python
import numpy as np

a = np.array([[1, 2], 
              [3, 4]])
b = np.array([[5, 6], 
              [7, 8]])

print(a+b)
print(a*b)
print("-----")
print(np.dot(a,b))
print(np.matmul(a,b))
print(a@b)

inverse = np.linalg.inv(a)
print(inverse)
print(np.linalg.det(a))
```
